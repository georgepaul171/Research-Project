```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/Users/georgepaul/Desktop/Research Project/Data Ideas/ComStock/Data/NY_baseline_metadata_and_annual_results (1).csv'

# Load the dataset
df = pd.read_csv(file_path)

# Basic Exploration
print("Head of the dataset:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nSummary Statistics (Numeric Columns):")
print(df.describe())

# Check for missing values in each column
print("\nMissing Values per Column:")
print(df.isna().sum())

# Identify numeric columns (continuous variables)
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("\nNumeric Columns:")
print(numeric_cols)

# Plot histograms for numeric columns
df[numeric_cols].hist(figsize=(15, 10))
plt.suptitle('Histograms of Numeric Variables', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Plot a correlation matrix using matplotlib's imshow
corr_matrix = df[numeric_cols].corr()
plt.figure(figsize=(12, 10))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation=90)
plt.yticks(range(len(corr_matrix)), corr_matrix.columns)
plt.title('Correlation Matrix of Numeric Variables', fontsize=16)
plt.tight_layout()
plt.show()
```
